```python
# Pregunta 5: Dibuje un gráfico de las acciones de Tesla

# Utilice la función make_graph para graficar los datos de acciones de Tesla y proporcione también un título para el gráfico.
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Get Tesla's stock data
tesla_data = yf.download('TSLA', start='2023-01-01', end='2023-12-31')

# Plot the 'Close' price over time
plt.figure(figsize=(12, 6))
plt.plot(tesla_data.index, tesla_data['Close'])
plt.title('Tesla Stock Price (2023)')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.grid(True)
plt.show()
```

```python
# Pregunta 6: Dibuje un gráfico de las acciones de GameStop

# Utilice la función make_graph para graficar los datos de acciones de GameStop y proporcione también un título para el gráfico.

```

```python
import yfinance as yf
import matplotlib.pyplot as plt

def make_graph(ticker, start_date, end_date, title):
    """
    Crea un gráfico de las acciones de una empresa específica.

    Args:
        ticker (str): El símbolo bursátil de la empresa.
        start_date (str): Fecha de inicio del período.
        end_date (str): Fecha de finalización del período.
        title (str): Título del gráfico.
    """

    # Obtener los datos de las acciones
    data = yf.download(ticker, start=start_date, end=end_date)

    # Graficar el precio de cierre
    plt.figure(figsize=(12, 6))
    plt.plot(data.index, data['Close'])
    plt.title(title)
    plt.xlabel('Fecha')
    plt.ylabel('Precio de Cierre (USD)')
    plt.grid(True)
    plt.show()

# Obtener los datos de GameStop
ticker = 'GME'
start_date = '2023-01-01'
end_date = '2023-12-31'
title = 'Precio de las Acciones de GameStop (2023)'

# Crear el gráfico
make_graph(ticker, start_date, end_date, title)
```
